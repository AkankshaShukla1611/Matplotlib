Introduction to the object oriented method

To begin we create a figure instance . then we can add axis to that figure

```python
#Create figure
fig = plt.figure()
#Add set of axes to figure    
axes = fig.add_axes([0.1,0.1,0.8,0.8,])
    #left,right,width,height,(range 0 to 1)
    
    # plot on that set of axes 
axes.plot(x,y,'b') #for blue color
axes.set_xlabel('Set X Label')  # set_to begin methods
axes.set_ylabel('Set Y Label')      
axes.set_title('Set Title')  
    
    
```

```python
# Create blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) #inset axes

# Larger Figure Axes 1
axes1.plot(x,y,'b')
axes1.set_xlabel(" x_label_axes1")
axes1.set_ylabel(" Y_label_axes1")
axes1.set_title(" title axes1")


# Insert Figure Axes 2
axes2.plot(y,x,'r')
axes2.set_xlabel(" x_label_axes2")
axes2.set_ylabel(" Y_label_axes2")
axes2.set_title(" title axes2")


```

subplots()


```python
# use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig , axes = plt.subplots()

# now use the axes object to add stuff to plot
axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

we can specify the no of rows and cols while creating the subplots() object

```python
# Empty canvas of 1 by 2 subplots
fig,axes = plt.subplots(nrows = 1,ncols = 2)
```

```python
# Axes is an array of axes to plot on axes
```

array([<Axes: >,<Axes : >], dtype = object)
We can iterate through this array:


```python
for ax in axes:
    ax.plot(x,y,'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
    
# Display the figure object
fig
```

A common issue with matplotlib is overlapping subplots or figures.
We can use fig.tight_layout() method , which automatically adjust the position 
of the axes on the figure canvas that there is no overlapping content:



```python
fig , axes = plt.subplots(nrows =1, ncols =2)
for ax in axes:
    ax.plot(x,y,'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
fig
# to make plots closer together
plt.tight_layout()
```

Matplotlib allows the aspect ratio, DPI and figure size to be specified 
when the Figure object is created. You can use the `figsize` and `dpi`
keyword arguments. 
* `figsize` is a tuple of the width and height of the figure in inches
* `dpi` is the dots-per-inch (pixel per inch). 

For example:

```python
fig = plt.figure(figsize=(8,4), dpi=100)
```

The same arguments can also be passed to layout managers, 
such as the subplots function:

```python
fig, axes = plt.subplots(figsize=(8,3))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```
